```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(ngram_range=(1,3), min_df = 0.001)

vectorized_text = vectorizer.fit_transform(mother['clean_text'])
vectorized_text = pd.DataFrame(
    vectorized_text.toarray(),
    columns = vectorizer.get_feature_names_out()
)

vectorized_text
```

```python
from sklearn.decomposition import LatentDirichletAllocation

# Instantiate the LDA
n_components = 50
lda_model = LatentDirichletAllocation(n_components=n_components, max_iter = 100, random_state=1, n_jobs = -1)

# Fit the LDA on the vectorized documents
lda_model.fit(vectorized_text)
```

```python
document_mixture = lda_model.transform(vectorized_text)
document_mixture.shape
```

```python
topic_word_mixture = pd.DataFrame(
    lda_model.components_,
    columns = vectorizer.get_feature_names_out()
)
topic_word_mixture
```

```python
def print_topics(lda_model, vectorizer, top_words):
    # 1. TOPIC MIXTURE OF WORDS FOR EACH TOPIC
    topic_mixture = pd.DataFrame(
        lda_model.components_,
        columns = vectorizer.get_feature_names_out()
    )

    # 2. FINDING THE TOP WORDS FOR EACH TOPIC
    ## Number of topics
    n_components = topic_mixture.shape[0]

    ## Top words for each topic
    for topic in range(n_components):
        print("-"*10)
        print(f"For topic {topic}, here are the the top {top_words} words with weights:")

        topic_df = topic_mixture.iloc[topic]\
            .sort_values(ascending = False).head(top_words)

        print(round(topic_df,3))
```

```python
print_topics(lda_model, vectorizer, 5)
```

```python
def get_topics(model, vectorizer):
    topics = []
    for idx, topic in enumerate(model.components_):
        topics.append([(vectorizer.get_feature_names_out()[i], topic[i])
                        for i in topic.argsort()[:-20 - 1:-1]])
    return topics

topics = get_topics(lda_model, vectorizer)
```
