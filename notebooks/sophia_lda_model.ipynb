```python
import pandas as pd
```

```python
top10_reviews = pd.read_csv('./csv/df_review_top10.csv')
top10_reviews['name'].unique()
```

```python
top10_reviews
```

```python
# Check the balance of the dataset
print(top10_reviews['stars'].value_counts())
```

```python
# Sample 3,000 reviews from each score category
balanced_top10_reviews = top10_reviews.groupby('stars', group_keys=False).apply(lambda x: x.sample(n=3000, random_state=1)).reset_index(drop=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(ngram_range=(2,3), min_df = 0.005)

vectorized_text = vectorizer.fit_transform(balanced_top10_reviews['clean_text'])
vectorized_text = pd.DataFrame(
    vectorized_text.toarray(),
    columns = vectorizer.get_feature_names_out()
)

vectorized_text
```

```python
from sklearn.decomposition import LatentDirichletAllocation

# Instantiate the LDA
n_components = 20 # try 20 number of topics
lda_model = LatentDirichletAllocation(n_components=n_components,
                                      max_iter=500,
                                      random_state=1,
                                      n_jobs=-1,
                                      doc_topic_prior=0.1,
                                      topic_word_prior=0.2)

# Fit the LDA on the vectorized documents
lda_model.fit(vectorized_text)
```

```python
document_mixture = lda_model.transform(vectorized_text)
document_mixture.shape
```

```python
topic_word_mixture = pd.DataFrame(
    lda_model.components_,
    columns = vectorizer.get_feature_names_out()
)
```

```python
def print_topics(model, vectorizer):
    for idx, topic in enumerate(model.components_):
        print("Topic %d:" % (idx))
        print([(vectorizer.get_feature_names_out()[i], topic[i])
                        for i in topic.argsort()[:-10 - 1:-1]])
```

```python
def get_topics(model, vectorizer):
    topics = []
    for idx, topic in enumerate(model.components_):
        topics.append([(vectorizer.get_feature_names_out()[i], topic[i])
                        for i in topic.argsort()[:-10 - 1:-1]])
    return topics
```

```python
topics = get_topics(lda_model, vectorizer)
topics
```

```python
import csv

# CSV file name
filename = "topics_scores.csv"

# Creating the CSV file
with open(filename, 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)

    # Writing the headers
    csvwriter.writerow(["Topic", "Phrase", "Score"])

    # Writing the data
    for topic_index, topic_list in enumerate(topics):
        for phrase, score in topic_list:
            csvwriter.writerow([topic_index, phrase, score])

print(f"CSV file '{filename}' created successfully.")
```

```python
import pandas as pd

def convert_topics_to_df(topics):
    columns = ["phrase", "score", "topic"]
    data = []  # To collect row data

    for idx, topic in enumerate(topics):
        for phrase, score in topic:
            data.append({
                "phrase": phrase,
                "score": score,
                "topic": idx
            })

    topics_df = pd.DataFrame(data, columns=columns)
    return topics_df


topics_df = convert_topics_to_df(topics)
topics_df
```
